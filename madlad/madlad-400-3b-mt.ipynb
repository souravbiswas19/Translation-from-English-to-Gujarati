```python
import os
file_path = "/kaggle/input/task-2/Task2.txt" #For Google Colab
#file_path = "Task2.txt" #For Google Colab
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()
```

```python
text
```

```python
import nltk
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

model_name = 'jbochi/madlad400-3b-mt'
model = T5ForConditionalGeneration.from_pretrained(model_name, device_map="auto")
tokenizer = T5Tokenizer.from_pretrained(model_name)
```

```python
def generate_gujarati(text):
    input_ids = tokenizer(text, return_tensors="pt", max_length=len(text), truncation=True).input_ids.to(model.device)
    outputs = model.generate(input_ids=input_ids)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
text_guj=[]
for sent in sentences:
  txt_guj = generate_gujarati(text=f"<2gu> {sent}")
  text_guj.append(txt_guj)
```

```python
for guj in text_guj:
    print(guj)
```

```python
text_in_gujarati = '\n'.join(text_guj)
print(text_in_gujarati)
```
