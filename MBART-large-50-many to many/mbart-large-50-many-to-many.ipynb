```python
import os
file_path = "/kaggle/input/task-2/Task2.txt" #For Google Colab
#file_path = "Task2.txt" #For Google Colab
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()
```

```python
text
```

```python
import nltk
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
article_en = "Hello, I am Sourav Biswas"
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
# translate English to Gujarati
```

```python
def generate_gujarati(text):
    tokenizer.src_lang = "en_XX"
    encoded_hi = tokenizer(text, return_tensors="pt")
    generated_tokens = model.generate(
        **encoded_hi,
        forced_bos_token_id=tokenizer.lang_code_to_id["gu_IN"]
    )
    translation=tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    return translation[0]
```

```python
text_guj=[]
for sent in sentences:
  txt_guj = generate_gujarati(text=sent)
  text_guj.append(txt_guj)
```

```python
for guj in text_guj:
    print(guj)
```
